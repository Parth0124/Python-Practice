```python
import matplotlib.pyplot as plt
```

```python
### create a line plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Basic line plot")
plt.show()
```

```python
### create a customize line plot

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red', linestyle='--', marker='o', linewidth=3,markersize=7)
plt.grid(True)
```
